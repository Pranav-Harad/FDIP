```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read grayscale image
img = cv2.imread(r"C:\Users\ML\Desktop\123B1F028\chittah.jpg", cv2.IMREAD_GRAYSCALE)
```

```python

# Gaussian noise parameters
mean = 0
sigma = 25  # standard deviation
gaussian = np.random.normal(mean, sigma, img.shape)

# Add Gaussian noise and clip values
noisy_img = img + gaussian
noisy_img = np.clip(noisy_img, 0, 255).astype(np.uint8)

# Plot original and noisy image with histograms
plt.figure(figsize=(12,6))

plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(noisy_img, cmap='gray')
plt.title("Gaussian Noise")
plt.axis('off')

plt.subplot(2,2,3)
plt.hist(img.ravel(), bins=256, range=[0,256], color='black')
plt.title("Histogram - Original")

plt.subplot(2,2,4)
plt.hist(noisy_img.ravel(), bins=256, range=[0,256], color='black')
plt.title("Histogram - Gaussian Noise")

plt.tight_layout()
plt.show()
```

```python
# Salt & Pepper noise parameters
prob = 0.02  # noise density
noisy_img = np.copy(img)

# Apply noise
rnd = np.random.rand(*img.shape)
noisy_img[rnd < (prob/2)] = 0        # pepper
noisy_img[rnd > 1 - (prob/2)] = 255  # salt

# Plot original and noisy image with histograms
plt.figure(figsize=(12,6))

plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(noisy_img, cmap='gray')
plt.title("Salt & Pepper Noise")
plt.axis('off')

plt.subplot(2,2,3)
plt.hist(img.ravel(), bins=256, range=[0,256], color='black')
plt.title("Histogram - Original")

plt.subplot(2,2,4)
plt.hist(noisy_img.ravel(), bins=256, range=[0,256], color='black')
plt.title("Histogram - Salt & Pepper Noise")

plt.tight_layout()
plt.show()
```

```python
# Impulse noise parameters
prob = 0.02  # noise density
noisy_img = np.copy(img)


mask = np.random.rand(*img.shape) < prob
random_values = np.random.randint(0, 256, img.shape, dtype=np.uint8)
noisy_img[mask] = random_values[mask]


plt.figure(figsize=(12,6))

plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(noisy_img, cmap='gray')
plt.title("Impulse Noise")
plt.axis('off')

plt.subplot(2,2,3)
plt.hist(img.ravel(), bins=256, range=[0,256], color='black')
plt.title("Histogram - Original")

plt.subplot(2,2,4)
plt.hist(noisy_img.ravel(), bins=256, range=[0,256], color='black')
plt.title("Histogram - Impulse Noise")

plt.tight_layout()
plt.show()
```

```python
# Speckle noise parameters
mean = 0
sigma = 0.2  # noise strength

# Generate Gaussian noise and multiply with image
gaussian = np.random.normal(mean, sigma, img.shape)
noisy_img = img + img * gaussian

# Clip values to valid range
noisy_img = np.clip(noisy_img, 0, 255).astype(np.uint8)

# Plot original and noisy image with histograms
plt.figure(figsize=(12,6))

plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(noisy_img, cmap='gray')
plt.title("Speckle Noise")
plt.axis('off')

plt.subplot(2,2,3)
plt.hist(img.ravel(), bins=256, range=[0,256], color='black')
plt.title("Histogram - Original")

plt.subplot(2,2,4)
plt.hist(noisy_img.ravel(), bins=256, range=[0,256], color='black')
plt.title("Histogram - Speckle Noise")

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imshow, imread
```

```python
img = cv2.imread(r'C:\Users\Pranav\OneDrive\Desktop\FDIP\Assignment1\mandala.jpg', cv2.IMREAD_GRAYSCALE) 
plt.imshow(img, cmap='gray')
plt.title("Original Grayscale Image")
plt.axis("off")
plt.show()
```

```python
# Increase brightness (+50)
brighter = cv2.convertScaleAbs(img, alpha=1, beta=50)

# Decrease brightness (-50)
darker = cv2.convertScaleAbs(img, alpha=1, beta=-50)

# Display brighter and darker images
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.imshow(brighter, cmap='gray')
plt.title("Brighter (+50)")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(darker, cmap='gray')
plt.title("Darker (-50)")
plt.axis("off")

plt.show()
```

```python
neg_img = 255 - img
title="Negative Transformation"
plt.imshow(neg_img, cmap='gray')
plt.title(title)
plt.show()
```

```python
img_float = img.astype(np.float32)

c = 255 / np.log(1 + np.max(img_float)) 
log_transformed = c * np.log(1 + img_float)

log_transformed = np.array(log_transformed, dtype=np.uint8)

plt.imshow(log_transformed, cmap='gray')
plt.title('Logarithmic Transformation')
plt.show()
```

```python
normalized_img = img / 255.0

gamma = 0.5  
c = 1.0        

gamma_corrected = c * np.power(normalized_img, gamma)

gamma_corrected = np.clip(gamma_corrected * 255, 0, 255).astype(np.uint8)

plt.imshow(gamma_corrected, cmap='gray')
plt.title(f'Gamma Correction ( = {gamma})')
plt.show()
```

```python
T = 150


_, binary_img = cv2.threshold(img, T, 255, cv2.THRESH_BINARY)


plt.imshow(binary_img, cmap='gray')
plt.title(f'Binary Thresholding (T = {T})')

plt.show()
```
